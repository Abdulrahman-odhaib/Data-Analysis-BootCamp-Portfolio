```python
import cv2
import numpy as np
from keras.models import model_from_json, Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input
from google.colab.patches import cv2_imshow

# خطوة 1: تحميل هيكل النموذج من ملف .json
with open("facial_expression_model_structure.json", "r") as json_file:
    model_json = json_file.read()

# خطوة 2: تعديل الكود باستخدام Input() لطبقة البداية
model_json = model_json.replace('"batch_input_shape": [null, 48, 48, 1]', '"input_shape": [48, 48, 1]')

# تحميل النموذج باستخدام model_from_json مع custom_objects
model = model_from_json(model_json, custom_objects={
    'Sequential': Sequential,
    'Conv2D': Conv2D,
    'MaxPooling2D': MaxPooling2D,
    'Flatten': Flatten,
    'Dense': Dense,
    'Input': Input
})

# خطوة 3: تحميل الأوزان من ملف .h5
model.load_weights('facial_expression_model_weights.h5')

# قائمة العواطف
emotion_labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

# تحميل الكاشف Haar لتحديد الوجوه
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# فتح الفيديو من الملف بدلاً من الكاميرا
cap = cv2.VideoCapture('download.jpeg')  # استبدل هذا بالمسار الفعلي للفيديو

# تحقق مما إذا كان الفيديو قد تم فتحه بنجاح
if not cap.isOpened():
    print("فشل في فتح الفيديو")
    exit()

while True:
    ret, frame = cap.read()

    if not ret:
        print("فشل في التقاط الإطار")
        break

    # تحويل الصورة إلى تدرجات الرمادي للكشف عن الوجوه
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # الكشف عن الوجوه
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # إذا تم العثور على وجوه، قم بمعالجتها
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # استخراج الوجه من الإطار
        face_region = frame[y:y + h, x:x + w]

        # تعديل حجم الوجه ليتوافق مع المدخلات في النموذج
        face_resized = cv2.resize(face_region, (48, 48))

        # تحويل الوجه إلى تدرجات الرمادي
        face_gray = cv2.cvtColor(face_resized, cv2.COLOR_BGR2GRAY)
        face_normalized = face_gray / 255.0

        # تعديل شكل البيانات ليتناسب مع المدخلات (1, 48, 48, 1)
        face_reshaped = np.reshape(face_normalized, (1, 48, 48, 1))

        # التنبؤ بالعاطفة (بدون @tf.function)
        predictions = model.predict(face_reshaped)
        predicted_emotion = emotion_labels[np.argmax(predictions)]

        # عرض العاطفة على الإطار
        cv2.putText(frame, predicted_emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # عرض الإطار مع العواطف
    cv2_imshow(frame)

    # الخروج إذا تم إغلاق النافذة
    if cv2.getWindowProperty('Emotion Detection', 0) < 0:
        break

# تحرير الفيديو بعد الانتهاء
cap.release()
```
